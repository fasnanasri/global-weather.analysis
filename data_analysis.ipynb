# DATA VISUALIZATION

# Bar Chart

- A bar chart helps to easily compare and analyze values across categories.

```python
import matplotlib.pyplot as plt

data = df['condition_text'].value_counts().head(10)

plt.figure(figsize=(10,6))
plt.bar(data.index, data.values, color='Blue')

plt.title('Weather Condition Frequency')
plt.xlabel('Condition')
plt.ylabel('Count')
plt.xticks(rotation=45)

plt.show()
```

- The bar chart shows the top 10 most frequent weather conditions in the dataset. It helps to identify which weather conditions occur most often and compare their frequencies easily.


```python
country_data = df['country'].value_counts().head(10)

plt.figure(figsize=(10,6))
plt.bar(country_data.index, country_data.values, color='green')

plt.title('Top Countries in Dataset')
plt.xticks(rotation=45)
plt.show()
```

- The bar chart shows the top 10 countries with the highest number of records in the dataset. It helps to compare which countries have more data entries than others.


# Simple Bar diagram

```python
import matplotlib.pyplot as plt

# Data
categories = ['A', 'B', 'C', 'D']
values = [10, 25, 15, 30]

# Create bar chart
plt.bar(categories, values, color='skyblue')

# Labels and title
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Simple Bar Diagram')

# Show the chart
plt.show()
```

- The bar chart compares values across different categories, showing which category has the highest and lowest values clearly.


# PIE CHART

```python
import matplotlib.pyplot as plt

# Sample data from your dataframe
cond = df['condition_text'].value_counts().head(5)
country = df['country'].value_counts().head(5)

plt.figure(figsize=(10,5))

# Plot 1 - Pie Chart
plt.subplot(1,2,1)
plt.pie(cond, labels=cond.index, autopct='%1.1f%%')
plt.title('Weather Conditions')

# Plot 2 - Bar Chart
plt.subplot(1,2,2)
plt.bar(country.index, country.values)
plt.title('Top Countries')
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

- The figure contains two charts: a pie chart showing the proportion of top weather conditions and a bar chart comparing the top countries by data count. It helps to understand both the distribution of weather conditions and the contribution of different countries in the dataset.


```python
import matplotlib.pyplot as plt

# Count top weather conditions
data = df['condition_text'].value_counts().head(6)

plt.figure(figsize=(7,7))
plt.pie(data, 
        labels=data.index, 
        autopct='%1.1f%%', 
        startangle=90)

plt.title('Top Weather Conditions')
plt.axis('equal')
plt.show()
```

- The pie chart shows the percentage distribution of the top 6 weather conditions. It helps to understand which weather conditions are most common and how each contributes to the total.

# Scatter Plot

##### Temperature vs Humidity

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8,5))

sns.scatterplot(x=gf["temperature_celsius"], y=gf["humidity"], color="Red")

plt.title("Humidity vs Temperature")
plt.xlabel("Temperature (°C)")
plt.ylabel("Humidity (%)")

plt.show()
```

- The scatter plot shows the relationship between temperature and humidity. It helps to observe patterns or trends, such as whether humidity increases or decreases with temperature.


##### Humidity vs PM2.5Air pollution

```python
plt.figure(figsize=(8,5))

sns.scatterplot(
    data=gf,
    x="humidity",
    y="air_quality_PM2.5",
    alpha=0.3
)

plt.title("Humidity vs PM2.5 Air Pollution")
plt.xlabel("Humidity (%)")
plt.ylabel("PM2.5")
plt.show()
```

- The scatter plot shows the relationship between humidity and PM2.5 air pollution levels. It helps to observe whether higher humidity is associated with higher or lower pollution levels and identify any patterns or trends.


# Histogram

### Temperature Distribution

```python
plt.figure(figsize=(8,5))

sns.histplot(df["temperature_celsius"], bins=30, kde=True, color="orange")

plt.title("Temperature Distribution")
plt.xlabel("Temperature (°C)")
plt.ylabel("Frequency")
plt.show()
```

- most temperature values are concentrated between 20°C  and 30°C.
- very low temperature (below 0°C ) and very high temperature (above 40°C) are rare.
- the data shows a peak around moderate temperatures ,indicating that extreme condition are less frequent.
